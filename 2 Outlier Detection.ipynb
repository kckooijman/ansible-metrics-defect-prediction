### Outlier Detection 

We use the following outlier detection models from the pyod package:

    Cluster-based Local Outlier Factor (CBLOF)
    Feature Bagging
    Histogram-base Outlier Detection (HBOS)
    Isolation Forest
    K Nearest Neighbors (KNN)
    Average KNN
    Local Outlier Factor (LOF)
    Minimum Covariance Determinant (MCD)
    One-class SVM (OCSVM)
    
First we scale the data using a MinMaxScaler, we combine the resulting outlier scores for the different outlier detection techniques using the outlier maximization score.

```python
import pandas as pd
import numpy as np
import math
import statistics
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager
from matplotlib import gridspec

#Outlier detection modules
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

from scipy import stats

import matplotlib.pyplot as plt
%matplotlib inline


from pyod.models.cblof import CBLOF 
from pyod.models.feature_bagging import FeatureBagging 
from pyod.models.hbos import HBOS 
from pyod.models.iforest import IForest 
from pyod.models.knn import KNN 
from pyod.models.lof import LOF
from pyod.models.mcd import MCD
from pyod.models.ocsvm import OCSVM
from pyod.models.combination import maximization
from pyod.utils.utility import standardizer


from IPython.display import display

pd.options.display.max_columns = None
```

We start by loading the data, and dropping the empty, and constant variables, as discovered in Notebook 1.

```python
data = pd.read_csv('results/output.csv')

filenames = data['file']
data = data.drop(columns = 'file')

to_drop = ['apls_count', 'niu_count', 'npl_count', 'npnn_count', 
           'npnn_count_relative', 'npun_count', 'npun_count_relative', 
           'npvr_count', 'nrgv_count', 'nrl_count', 'nrvr_count', 
           'nvr_count', 'nscm_count', 'nmo_count', 'nimpp_count']

data = data.drop(columns=to_drop)
data = data.dropna()
print("Data shape after dropping na columns and rows: ", data.shape)
data.head()
```

```python
random_state = np.random.RandomState(0)

scaler = MinMaxScaler(feature_range=(0,1))
X = scaler.fit_transform(data)
# pca = PCA(n_components=2, random_state=random_state)
# pca.fit(data_scaled)
# X = pca.transform(data_scaled)


outliers_fraction = 0.05
classifiers = {     
    'Cluster-based Local Outlier Factor (CBLOF)':
        CBLOF(contamination=outliers_fraction,
              check_estimator=False, random_state=random_state),
    'Feature Bagging':
        FeatureBagging(LOF(n_neighbors=35),
                       contamination=outliers_fraction,
                       random_state=random_state),
    'Histogram-base Outlier Detection (HBOS)': HBOS(
        contamination=outliers_fraction),
    'Isolation Forest': IForest(contamination=outliers_fraction,
                                random_state=random_state),
    'K Nearest Neighbors (KNN)': KNN(
        contamination=outliers_fraction),
    'Average KNN': KNN(method='mean',
                       contamination=outliers_fraction),
    'Local Outlier Factor (LOF)':
        LOF(n_neighbors=35, contamination=outliers_fraction),
    'Minimum Covariance Determinant (MCD)': MCD(
        contamination=outliers_fraction, random_state=random_state),
    'One-class SVM (OCSVM)': OCSVM(contamination=outliers_fraction),
}
```

```python
scores = np.zeros([X.shape[0], len(classifiers)])
for i, (clf_name, clf) in enumerate(classifiers.items()):
    try:
        clf.fit(X)
        scores[:, i] = clf.decision_scores_
    except:
        print("Failed for ", clf_name)
    
scores_norm = standardizer(scores)
comb_max = maximization(scores_norm)
data['outlier_scores'] = comb_max
data['outlier'] = [1 if x>3 else 0 for x in data['outlier_scores']]
data['file'] = filenames
data.to_csv("results/results_with_outliers.csv", index=False)
```

```python
plt.plot(sorted(data['outlier_scores']))
plt.xlabel('Number of files')
plt.ylabel('Outlier maximization score')
plt.hlines(y=3, xmin=0, xmax=9400, color="red")

plt.savefig("figures/outlier_maximization_score.png", dpi=300)
plt.show()
```

In the above plot, we find an outlier-cutoff for y=3. This gives us a total of 397 outliers.

## Testing significance for difference between outliers and non-outliers per metric

```python
from decimal import Decimal

def ks(data, varname):
    return stats.ks_2samp(list(data[data['outlier']==0][varname]),
                   list(data[data['outlier']==1][varname]))

metrics = data.columns[:-2]
significant = [m for m in metrics if ks(data, m)[1]<=0.05]
unsignificant = [m for m in metrics if ks(data, m)[1]>0.05]
for m in metrics:
    res = ks(data, m)
    inliers = data[data['outlier']==0][m]
    outliers = data[data['outlier']==1][m]
    #print("{}: significant: {}, p: {},".format(m, res[1]<0.05, res[1]))
    print(m.replace("_count", ""), "&", round(statistics.mean(inliers),2), "&", round(statistics.mean(outliers),2), "&", res[1]<0.05, "&", "{:.2e}".format(Decimal(res[1])), "\\\\")
```

```python
fig = plt.figure(figsize=(10,50))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
fig.text(0.04, 0.5, 'density', va='center', rotation='vertical')
for i in range(len(significant[:40])):
    ax = fig.add_subplot(20,5,i+1)
    plt.hist(data[data['outlier']==0][significant[i]], alpha=0.5, label='inliers', density=True)
    plt.hist(data[data['outlier']==1][significant[i]], alpha=0.5, label='outliers', density=True)
    ax.set_title(significant[i].replace("_count", ""))
    
plt.show()
```
